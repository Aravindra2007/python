```python
# logical operator

print(1 or 5)
print(1 and 5)  # for and botth should be true otherwise it will print 0  or max number
print(0 or 5)
print(0 and 5)

print(True and False)
print(True and True)
print(True or False)
print(False or False)
```

```python
# Bitwise operators
print(4 >> 1)  # it is shifted to right side one bit
print(4 and 3)  # in this bitswill be considered and if both bits contain one then it takes one other wise zero
print(4 or 3) # either one bit should be true from both numbers
print(~3)  # compliment will be done here
print(3 ^ 4) # both bits should be different then it is 1
print(4 << 2) # 2 bits is moved towards the left side 
```

```python
## comparsion operator
x, y = 10, 20
print('--- Basic Comparisons ---')
print(f'{x} == {y}  -> {x == y}')   # Equal
print(f'{x} != {y}  -> {x != y}')   # Not equal
print(f'{x} >  {y}  -> {x > y}')    # Greater than
print(f'{x} <  {y}  -> {x < y}')    # Less than
print(f'{x} >= {y}  -> {x >= y}')   # Greater or equal
print(f'{x} <= {y}  -> {x <= y}')   # Less or equal

print()
```

```python
## Assignment Operators 
## after appling this operators and again stored in the same variable
x = 10
print(f'Initial x = {x}')

x += 5;  print(f'After x += 5  : {x}')   # addition 
x -= 3;  print(f'After x -= 3  : {x}')   # minus
x *= 2;  print(f'After x *= 2  : {x}')   # multipli
x /= 4;  print(f'After x /= 4  : {x}')   # divison it can be float points
x //= 2; print(f'After x //= 2 : {x}')   # floor diviso
x **= 3; print(f'After x **= 3 : {x}')   # power should multiplied two times
x %= 5;  print(f'After x %%= 5  : {x}')  # modules will be done two times   

print()
```

```python
## Memborship opeartor (in and not in)
sentence = 'Python is awesome for data science'
print(f'"Python" in sentence     -> {"Python" in sentence}')
print(f'"Java" in sentence       -> {"Java" in sentence}')
print(f'"Java" not in sentence   -> {"Java" not in sentence}')
print(f'"data" in sentence       -> {"data" in sentence}')

print()

# ── With Lists ──
print('--- Membership in Lists ---')
fruits = ['apple', 'banana', 'mango', 'cherry', 'grape']
for item in ['mango', 'pineapple', 'banana', 'watermelon']:
    if item in fruits:
        print(f'  ✅ "{item}" is available')
    else:
        print(f'  ❌ "{item}" is NOT in stock')

print()
```
