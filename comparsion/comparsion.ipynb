```python
#Basic Operator / Arthematic operator
a = 30
b =10
print("Addition : ",a + b)
print("subtraction : ",a - b)
print("Multiplication : ",a*b)
print("Division : ",a/b)
print("Floor Division : ",a//b)
print("remainder : ",a%b)
```

```python
number = int(input("Enter a number: "))
if number % 2==0:
    print(f"{number} is an even number.")
else:
    print(f"{number} is an odd number.")
```

```python
# logical operator

print(1 or 5)
print(1 and 5)  # for and botth should be true otherwise it will print 0  or max number
print(0 or 5)
print(0 and 5)
```

```python
# Bitwise operators
print(4 >> 1)  # it is shifted to right side one bit
print(4 and 3)  # in this bitswill be considered and if both bits contain one then it takes one other wise zero
print(4 or 3) # either one bit should be true from both numbers
print(~3)  # compliment will be done here
print(3 ^ 4) # both bits should be different then it is 1
print(4 << 2) # 2 bits is moved towards the left side 
```
