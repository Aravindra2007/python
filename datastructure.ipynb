```python
## List Operations 
l =[1,3,5,9,7,4,7,1]

print("Length of a list: ",(len(l)))

print("Last element will be popped : ",l.pop())

print("2nd index element will be popped : ",l.pop(2))

l.sort()   
print("Sorted list : ",l)

l.insert(6,34)      #inserting the element with index
print("inserting the element with index : ",l)

ls = [70,80]
l.extend(ls)  ## is used to merge the two lists
print("After merging the two lists : ",l)

l.remove(70)        # it will remove the particular element mentioned
print("list after removing the element 70 : ",l)


print(l.index(34))  # gives the particular element index

print("count is : ",l.count(7))   # it counts the particular element 

l.sort(reverse=True)  # descending order of the given elements 
print("descending  order : ",l)

l.reverse()
print("Reverse order : ",l)

copied = l.copy()
print("Copied list : ",copied)

copied.clear()
print("list will be cleared : ",copied)
```
