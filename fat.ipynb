```python
def even(t):
    if not t:
        return 0
    sum = t[0]
    if sum%2 == 0:
        return sum + even(t[1:])
    else:
        return even(t[1:])
    
numbers = (1, 2, 3, 4, 5, 6, 7, 8)
print(even(numbers))
    
```

```python
def ser(n,x):
    if n <= 0:
        return 1
    else:
        sum = ((-1)**(n+1)) * (x**(n-1))
        return sum + ser(n-1,x)
    
print(1+ser(3,2))
```

```python
A = [[1, 2],
     [3, 4]]

B = [[5, 6],
     [7, 8]]

C = [[0, 0],
     [0, 0]]

for i in range(len(A)):
    for j in range(len(A)):
        C[i][j] = A[i][j] + B[i][j]

print("Addition:")
for row in C:
    print(row)
```

```python
A = [[1, 2],
     [3, 4]]

B = [[5, 6],
     [7, 8]]

C = [[0, 0],
     [0, 0]]


for i in range(len(A)):
    for j in range(len(B)):
        for k in range(len(B)):
            C[i][j] += A[i][k] * B[k][j] 

print(C)
```
