```python
import re
text = "hello wolrd welcome to python programming  age 18 now 20"
pat = r'hello'

result = re.findall(pat,text)
result1 = re.match(pat,text)
print(result)
print(result1)
```

```python
pat = r'\d{10}'
result = re.findall(pat,text)
result2 = re.search(pat,text)
print(result2)
print(result)
```

```python
l = "hai"
pat = r'h...o'
res = re.sub(pat,l,text)
print(res)
```
