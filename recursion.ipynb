# Recursion
  -> it is a technique where a function call itself to solve amaller version of the same problem 
  -> By defaulting it recursion to 1000 calls by default  (sys.setrecursionlimit(n))

```python
def fact(num):
    if num == 1:
        return 1
    else:
        return num * fact(num-1)

print(fact(5))
```

```python
num = 10

b = bin(num)
x = hex(num)
o = oct(num)
print(b)
print(x)
print(o)
```

```python
def fibo(num):
    a,b = 0,1
    print(a,end=" ")
    count = 0
    while count < num-1:
        print(b,end=" ")
        a,b = b,a+b
        count += 1
        

fibo(10)
```

```python
def binary(num):
    if num == 0:
        return 1
    else:
        x = num%2
        num = num/2
        return binary(num)
    
print(binary(10))
```

```python
def binary1(num):
    rev = ""
    while num > 0:
        rev = str(num%2) + rev
        num = num//2
    return rev
        

binary1(10)
```
