### 1. LISTS []
What is a List?
A List is an ordered, mutable (changeable), and indexed collection of items. Lists allow duplicate values and can hold items of different data types (integers, strings, floats, even other lists).

Key Characteristics:
Property:	        Value:
Ordered	            ✅ Yes
Mutable	            ✅ Yes
Allows Duplicates	✅ Yes
Indexed	            ✅ Yes (0-based)
Syntax	            [ ]

<!-- Method	Description -->
append(x)	            Adds item x at the end
insert(i, x)	        Inserts x at index i
extend(iterable)	    Adds all items from iterable
remove(x)	            Removes first occurrence of x
pop(i)	                Removes & returns item at index i (default: last)
clear()	                Removes all items
index(x)	            Returns index of first occurrence of x
count(x)	            Returns count of x in list
sort()	                Sorts the list in-place
reverse()	            Reverses the list in-place
copy()	                Returns a shallow copy
len(list)	            Returns number of elements (built-in)


```python
## List Operations 
l =[1,3,5,9,7,4,7,1]

print("Length of a list: ",(len(l)))

print("Last element will be popped : ",l.pop())

print("2nd index element will be popped : ",l.pop(2))

l.sort()   
print("Sorted list : ",l)

l.insert(6,34)      #inserting the element with index
print("inserting the element with index : ",l)

ls = [70,80]
l.extend(ls)  ## is used to merge the two lists
print("After merging the two lists : ",l)

l.remove(70)        # it will remove the particular element mentioned
print("list after removing the element 70 : ",l)


print(l.index(34))  # gives the particular element index

print("count is : ",l.count(7))   # it counts the particular element 

l.sort(reverse=True)  # descending order of the given elements 
print("descending  order : ",l)

l.reverse()
print("Reverse order : ",l)

copied = l.copy()
print("Copied list : ",copied)

copied.clear()
print("list will be cleared : ",copied)
```

### 2. TUPLES ()
What is a Tuple?
A Tuple is an ordered, immutable (unchangeable), and indexed collection. Once created, its elements cannot be modified. Tuples are faster than lists and are used to store data that should not change.

<!-- Key Characteristics: -->

Ordered	            ✅ Yes
Mutable	            ❌ No (Immutable)
Allows Duplicates	✅ Yes
Indexed	            ✅ Yes
Syntax	            ( )

<!-- 2.2 Tuple Methods -->
Tuples are immutable, so they have only 2 built-in methods:

<!-- Method	Description -->
count(x)	        Returns how many times x appears
index(x)	        Returns index of first occurrence of x
Additionally,       these built-in functions work on tuples:

<!-- Function Description -->
len(t)	    Number of elements
min(t)	    Minimum value
max(t)	    Maximum value
sum(t)	    Sum of all elements
sorted(t)	Returns a sorted list (doesn't modify tuple)

```python
# ── 🎯 Practical Program: GPS Coordinate Tracker ─────────────────

print("=" * 45)
print("       GPS COORDINATE TRACKER")
print("=" * 45)

# Storing locations as tuples (immutable — coordinates don't change!)
locations = {
    "Home"      : (17.3850, 78.4867),
    "College"   : (17.4400, 78.4983),
    "Hospital"  : (17.3616, 78.4747),
    "Airport"   : (17.2403, 78.4294),
}

import math

def distance(loc1, loc2):
    """Approximate distance in km between two lat/lon points."""
    lat1, lon1 = loc1
    lat2, lon2 = loc2
    return math.sqrt((lat2 - lat1)**2 + (lon2 - lon1)**2) * 111

print("\n📍 Saved Locations:")
for name, coords in locations.items():
    lat, lon = coords   # tuple unpacking
    print(f"  {name:<12}: Lat={lat}, Lon={lon}")

home = locations["Home"]
print("\n📏 Distances from Home:")
for name, coords in locations.items():
    if name != "Home":
        d = distance(home, coords)
        print(f"  Home → {name:<12}: {d:.2f} km")
```

### Sets {}

A Set is an unordered, mutable, collection of unique elements. Sets do not allow duplicates and have no index. They are ideal for membership testing and eliminating duplicates.

<!-- Key Characteristics: -->
<!-- Property	Value -->
Ordered	            ❌ No
Mutable	            ✅ Yes
Allows Duplicates	❌ No
Indexed	            ❌ No
Syntax	            { }


<!-- All Set Methods — Explained -->
Modification Methods:
Method	                Description
add(x)	                Adds element x
update(iterable)	    Adds multiple elements
remove(x)	            Removes x; raises error if not found
discard(x)	            Removes x; no error if not found
pop()	                Removes and returns an arbitrary element
clear()	                Removes all elements
<!-- Set Operations (Mathematical): -->
Method	            Operator	        Description

union(s)	            \|	                All elements from both sets
intersection(s)	        &	                Elements common to both
difference(s)	        -	                Elements in first but not second
symmetric_difference(s) ^	                Elements in either but not both
issubset(s)             <=	                Is this set a subset?
issuperset(s)	        >=	                Is this set a superset?
isdisjoint(s)	        —	                Do sets share no elements?

### DICTIONARIES {key : value}
What is a Dictionary?
A Dictionary stores data as key-value pairs. It is ordered (Python 3.7+), mutable, and keys must be unique and immutable (strings, numbers, tuples). Think of it like a real dictionary — you look up a word (key) to get its meaning (value).

<!-- Key Characteristics: -->
Property	            Value
Ordered	                ✅ Yes (Python 3.7+)
Mutable	                ✅ Yes
Duplicate Keys	        ❌ No
Duplicate Values	    ✅ Yes
Syntax	                { key: value }

 <!-- All Dictionary Methods — Explained -->
Method	                                        Description
get(key, default)	                    Returns value for key; default if not found
keys()	                                Returns all keys
values()	                            Returns all values
items()	                                Returns all key-value pairs as tuples
update(dict2)	                        Merges dict2 into the dictionary
pop(key)	                            Removes key and returns its value
popitem()	                            Removes and returns the last inserted item
setdefault(key, val)	                Returns value; sets it if key doesn't exist
copy()	                                Returns a shallow copy
clear()	                                Removes all key-value pairs
fromkeys(keys, val)	                    Creates dict from keys with a default value
