loop controls are 

break ->  it will stop the processing and ends the iteration
continue -> it will skip the that current value
pass -> it will continue the all value (even though our condition is true but we dont wanna do ntg so we can use pass)


```python
num = int(input("Enter the Number : "))
for i in range(num):
    if i%2 == 0:
        continue
    else:
        print(i)
```

```python
# print the 1st 5 prime numbers from the given number 
num = int(input("enter the number : "))
count = 0
while count < 5:
    prime = True
    if num < 2:
        prime = False
    else:
        for i in range(2,num):
            if num % i == 0:
                prime = False
                break
    if prime:
        print(num)
        count += 1

    num += 1
```

```python
## printing the prime number from given number upto five terms

num = int(input("Enter the Number : "))
count = 0
while count < 5:
    prime = True
    if num < 2:
        prime = False
    else:
        for i in range(2,num):
            if num%i == 0:
                prime = False
                break
    if prime:
        print(num)
        count += 1

    num += 1
```

```python
## break statemnt example 
num = int(input("enter the Number : "))

for i in range(1,3):
    for j in range(1,5):
        if j > 3:
            break
        print(j)
```

```python
## continue Statement
# numbers from 1 to num and skip multiple of 3

num = int(input("Enter the Number ; "))
for i in range(1,num+1):
    if i%3 == 0:
        continue    
    print(i,end=" ")
```

```python
## pass (it will make nothing but the purpose we can write or replace in the place of pass in future so it will be helpfull by doing simply nothing )
num = int(input("Enter the number :"))
for i in range(1,num):
    if i%2 == 0:
        pass
    print(i,end=" ")
    
```

```python
## combination break and continue 
# processing the numbers btw 1-15 skip multiple of 3 and stop at 13
num = int(input("Enter the number : "))
for i in range(1,num):
    if i > 13:
        break
    else:
        if i%3 == 0:
            continue
        print(i,end=" ")
    
```

```python
## break with nested loops  - break only exists innermost loop
for i in range(1,4):
    print("outer loop ",i)
    for j in range(1,4):
        if j == 3:
            print("Breaks at : ",j)
            break
        print("inner loop : ",j)
```
