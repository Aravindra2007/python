### Comman pattern Syntax 
### pattern   and meaning 
## r should be mandotory if we use the \ slash symbol other ise it didnt consider it as regular expression(ex : pat = r"\d+"  )

. - > any charcter except newline
^ ->  start of string
$ -> end of string/text
* -> 0 or more repetitions
+ -> 1 or more repetitions 
? -> 0 or 1 repetition
{n} -> exactly n repetitions 
{m},{n} between m and n repetitions
[]-> character set 
| -> or operator
() -> grouping 
\d -> any digit (0-9)
\D -> any non-digit
\w -> word character (a-z,A-Z,0-9,_)
\W -> non word character

```python
import re
text = "the price is $14.99 and the discount is $5.00"
match = re.search(r"\$\d+\.\d{2}",text)            ## for giving the + it gives all values 


print(bool(match))
print(bool(match.group))
print(match.start(),match.end())
```

```python
text = "The price is $49.99 and the discount is $5.00"
match = re.search(r"\$\d+\.\d{2}", text)
if match:
    print(f"First price found: {match.group()}")
    print(f"Position: {match.start()} to {match.end()}")
```
