```python
amount = int(input())
count = 0
while amount > 0:
    if amount >= 500:
        digit = amount//500
        amount = amount%500
        count += digit
        print("500 notes : ",digit)
    elif amount >= 100:
        digit = amount //100
        amount = amount%100
        count += digit
        print('100 notes : ',digit)
```
