```python
## enumerate 
msg = "hello hai"
for idx,ch in enumerate(msg):
    print(f"index  {idx} char {ch}")
```

```python
## printing the character starting the index by using the start function
for idx,ch in enumerate(msg,start=1):
    print(f"{idx} char {ch}")
```

```python
### zip()

student = ["ak","bk","ck"]
marks = [97,92,89]
for s,m in zip(student,marks):
    print(f"student {s} marks {m}")
```

```python
## printing the sting reverse using the keyword reversed
msg = "hai hello"
for i in reversed(msg):
    print(i,end=" ")
print()
for i in reversed(range(len(msg))):
    print(msg[i],end=" ")
```
