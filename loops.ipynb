```python
# Traingle Using for loops 
#right angled triangle
n = 5
for i in range(1,n+1):
    print("* "*i)
    
```

```python
# reverse right angle triangle
for i in range(n,0,-1):
    print("* "*i)
```

```python
## Full Pyiramids
for i in range(1,n+1):
    print(" "*(n-i)+"* "*i)
```

```python
## inverted pyramids
for i in range(n,0,-1):
    print(" "*(n-i)+"* "*i)
```

```python
## diamond Shape
for i in range(1,n+1):
    print(" "*(n-i)+"* "*i)
for i in range(n-1,0,-1):
    print(" "*(n-i)+"* "*i)
```

```python
## right angle triangle with numbers
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
## invert right angle triangle with numbers
for i in range(n,0,-1):
    for j in range(i,0,-1):
        print(j,end=" ")
    print()
```

```python
## armstrong Number
num = int(input("enter the number "))
sum = 0
while num > 0:
    digit = num%10
    sum += digit *digit*digit
    num = num//10
        
if(sum == num):
    print("Armstrong Number ")
```
